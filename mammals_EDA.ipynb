```python
# Importing required libraries
import pandas as pd
import numpy as np
import seaborn as sn
import matplotlib as mpl
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
```

```python
# Open dataset
mammals_df = pd.read_csv('Resources/mammals.csv')
mammals_df.head()
```

```python
# columns
mammals_df.columns.to_list()
```

```python
mammals_df.info()
```

```python
# find out the unique value to make categorical variable
print('Species')
print(*list(mammals_df.species.unique()))
print('Body Weight')
print(*list(mammals_df.body_wt.unique()))
print('Brain Weight')
print(*list(mammals_df.brain_wt.unique()))
print('Non-dreaming')
print(*list(mammals_df.non_dreaming.unique()))
print('Total Sleep')
print(*list(mammals_df.total_sleep.unique()))
print('Life Span')
print(*list(mammals_df.life_span.unique()))
print('Gestation')
print(*list(mammals_df.gestation.unique()))
print('Predation')
print(*list(mammals_df.predation.unique()))
print('Exposure')
print(*list(mammals_df.exposure.unique()))
print('Danger')
print(*list(mammals_df.danger.unique()))
```

```python
# Looking at totals 
mammals_df.describe()
```

```python
# Looking at NAN
mammals_df.isnull().sum()
```

```python
#Visual Analysis of dreaming 
sn.factorplot(x='dreaming', kind='count', data=mammals_df, height=8, aspect=15/8)
```

```python
#Visual Analysis of non-dreaming 
sn.factorplot('non_dreaming', kind='count', data=mammals_df, height=8, aspect=15/8)
```

```python
#heatmap 
corr=mammals_df.corr(method='spearman')
plt.figure(figsize=(15,15))
sn.heatmap(corr,vmax=.8,linewidths=0.01,square=True,annot=True,cmap='YlGnBu',linecolor='black')
```

```python
#dreaming Vs predation
b = sn.barplot(x=mammals_df.dreaming, y=mammals_df.predation)

#Increasing font size
sn.set(font_scale=3)
b.set_xlabel("Dreaming", fontsize = 30)
b.set_ylabel("Predation", fontsize = 30)
```
